```python
import pandas as pd
import matplotlib as mp
import seaborn as sns
```

```python
epochs = pd.read_csv("epoch_stat.csv", delimiter=',')
epochs.head()
```

```python
epochs.plot(x='epoch', y='loss', kind='line',figsize=(15,8), title='(final) loss over 100 epochs')
```

```python
iteration = pd.read_csv("iteration_stat.csv", delimiter=',')
iteration.head()
```

```python
iteration[iteration.epoch==0].plot(x='iteration', y='loss', kind='line',figsize=(15,8),title='Loss over 0th epoch')
```

```python
iteration[iteration.epoch<10].plot(x='iteration', y='loss', kind='line',figsize=(15,8),title='Loss over 10 first epochs')
```

```python
it_epoch_group = iteration.groupby("epoch")["loss"].mean()
```

```python
it_epoch_group.plot(x='epoch', y='loss', kind='line',figsize=(15,8),title='Average loss', xlim=(0,7))
```
